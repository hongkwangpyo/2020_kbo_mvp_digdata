```python
url = 'http://www.kleague.com/rank?ch=023806'
res = requests.get(url)
bs = BeautifulSoup(res.text,'html.parser')
```

```python
lis = bs.select('#responsive_on > nav > div.nav-top > div > div > div.nav-league-part.clearfix > div.selector.float-left > div > ul > li')
k_leage_team = {}
for li in lis:
    href = li.select('a')[0]['href']
    team = li.select('a')[0]['title']
#     print(href)
    k_leage_team[team] = href.split('=')[1]

print(k_leage_team)
```

```python
url = 'http://www.kleague.com/club?club={}'
count = 0
player_data = {}
for team,code in k_leage_team.items():
    if count >= 3:
        break
    url = url.format(code)
    res = requests.get(url)
    bs = BeautifulSoup(res.text,'html.parser')
    divs = bs.select('#pills-members > div.members.row > div')
    for div in divs:
        position = div['class'][2].split('_')[1]
        player_id = div['id']
        if position != '':
            player_data[player_id] = player_id
```
